```python
import os
import re # 데이터 정제 패키지
import cv2
import numpy as np

test_image = "c:\\image"

def image_load(path):
    file_list = os.listdir(path)
    #return file_list
    file_name = []
    for i in file_list:
        a = int( re.sub('[^0-9]','',i) ) # 숫자가 아닌 것은 null 처리
        file_name.append(a)
    file_name.sort()
    for j in range(len(file_list)):
        file_name[j] = path+'\\'+str(j+1)+'.png' 
    
    image = []
    for k in file_name:
        img = cv2.imread(k) # 파일을 읽어서 숫자 행렬로 변환
        image.append(img) 
        
    return np.array(image) # 숫자 행렬 리스트를 numpy array로 감싼다.

print(image_load(test_image))
```

```python
import pandas as pd

df = pd.read_csv('d:\\data\\csv\\train.csv')

def missing_values(x):
    return x.isnull()

missing_values(df)

def missing_count(x):
    return missing_values(x).sum()

missing_count(df)

def total_number_missing(x):
    return missing_count(x).sum()

total_number_missing(df)
```

```python

df.pipe(missing_count)
```

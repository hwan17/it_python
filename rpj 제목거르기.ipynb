```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
#from urllib import parse


def mg(month):
    #keyword = '+'+parse.quote(key)
    #print(keyword)
    
    if month < 10:
        month ='0'+str(month)
    else:
        month = str(month)
        
    keyls = ['%C4%DA%B7%CE%B3%AA+%BA%F1%B4%EB%B8%E9','%C4%DA%B7%CE%B3%AA+%BE%F0%C5%C3%C6%AE','%C4%DA%B7%CE%B3%AA+%BF%C2%B6%F3%C0%CE','%C4%DA%B7%CE%B3%AA+%BF%F8%B0%DD','%C4%DA%B7%CE%B3%AA+%B7%A3%BC%B1']
    # 코로나 + 비대면 언택트 온라인 랜선 원격
    #month =
    svurl = []
    for kwd in keyls:
        for pn in range(1,1000000):
            list_url = "http://find.mk.co.kr/new/search.php?pageNum="+str(pn)+"&cat=&cat1=&media_eco=&pageSize=10&sub=all&dispFlag=OFF&page=news&s_kwd="+kwd+"&s_page=news&go_page=&ord=1&ord1=1&ord2=0&s_keyword="+kwd+"&period=p_direct&s_i_keyword="+kwd+"&s_author=&y1=2020&m1="+month+"&d1=01&y2=2020&m2="+month+"&d2=31&ord=1&area=ttbd"
            #print(list_url)
            url = urllib.request.Request(list_url)
            print(url)
            #try:
            result = urllib.request.urlopen(url).read().decode("cp949")
            soup = BeautifulSoup(result,'html.parser')
            result1 = soup.find_all('span',class_="class_tit")
            artn=result1[0].get_text()
            n=artn.index('건')
            tnum=artn[11:n].replace(',','')
            en = int(int(tnum)/20)+1
            #print(en)
            if pn == en+1:
                break
            print(artn)
            result2 = soup.find_all('span',class_="art_tit")
            for j in result2:
                for k in j:
                    #print(k.get("href"))
                    svurl.append(k.get("href"))
       # except:
        #        pass
    print(len(svurl))
    print(len(list(set(svurl))))
    return svurl

#-------

def mg_scrl(month):
    #params=[]
    tmpmg = mg(month)
    params1=[]

    for ulst in list(set(tmpmg)):
        print(ulst)
        #list_url = 
        url = urllib.request.Request(ulst)
        result = urllib.request.urlopen(url).read()#.decode("cp949")
        #print(result)
        soup = BeautifulSoup(result, "html.parser")
        
        resultt = soup.find_all('h1', class_='top_title')
        print(resultt)
        resultst = soup.find_all('h2', class_='sub_title1')
        print(resultst)
        tt = resultt+resultst
        print(tt)
        esccnt = 0
        for kw in ['코로나','비대면','언택트','온라인','랜선','원격','오프라인']:
            if tt.count(kw) != 0:
                esccnt += 1
                break
        if esccnt != 0 :
               
            result1 = soup.find_all('div', class_='art_txt')
            #print(result1)
            #result2 = soup.find_all('p', class_="con")

            #params2=[]
            for i in result1:
                params1.append(i.get_text(" ",strip=True)+'\n'+'\n')

        #for i in result2:
        #    params2.append(i.get_text(" ",strip=True))

        #for i,j in zip(params1,params2):
        #    params.append(i + ' ' + j+'\n')
#     f = open('c:\\data\\'+'crl'+str(month)+'.txt','w',encoding = "utf-8")
#     #print(params1)
#     f.writelines(params1)
#     f.close()
    #return params1
    param = pd.DataFrame(params1)
    param.to_csv('d:\\data\\R_pj\\'+'tcrl'+str(month)+'.csv', encoding='utf-8')
```

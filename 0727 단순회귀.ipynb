```python
# -*- coding: utf-8 -*-

###   

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
'''

[Step 1]   - read_csv()     

'''

# CSV   

df = pd.read_csv('D:\\data\\csv\\\\auto-mpg.csv', header=None)

#   

df.columns = [
    'mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
    'model year', 'origin', 'name'
]

#  

print(df.head())

print('\n')

# IPython   -     

pd.set_option('display.max_columns', 10)

print(df.head())

print('\n')
'''

[Step 2]  

'''

#   
# horsepower   object    .
print(df.info())

print('\n')

#    

print(df.describe())

print('\n')

# horsepower    ( ->)

print(df['horsepower'].unique())  # horsepower   

print('\n')

df['horsepower'].replace('?', np.nan, inplace=True)  # '?' np.nan 

df.dropna(subset=['horsepower'], axis=0, inplace=True)  #   

df['horsepower'] = df['horsepower'].astype('float')  #   

print(df.describe())  #    

print('\n')
'''

[Step 3] (feature  variable) 

'''

#   ()  (, , , )

ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]

print(ndf.head())

print('\n')

###   Y "(mpg)"     () 

# Matplotlib  

ndf.plot(kind='scatter', x='weight', y='mpg', c='coral', s=10, figsize=(10, 5))
#  gallon    mile   gallon    mile .
plt.show()

plt.close()

# seaborn  

fig = plt.figure(figsize=(10, 5))

ax1 = fig.add_subplot(1, 2, 1)

ax2 = fig.add_subplot(1, 2, 2)

sns.regplot(x='weight', y='mpg', data=ndf, ax=ax1)  #  

sns.regplot(x='weight', y='mpg', data=ndf, ax=ax2, fit_reg=False)  # 

plt.show()

plt.close()

# seaborn   - , 

sns.jointplot(x='weight', y='mpg', data=ndf)  #  

sns.jointplot(x='weight', y='mpg', kind='reg', data=ndf)  #  

plt.show()

plt.close()

# seaborn pariplot       

sns.pairplot(ndf)

plt.show()

plt.close()
'''

Step 4:   - (train data)/ (test data)

'''

# () 

X = ndf[['weight']]  #  X

y = ndf['mpg']  #  Y

# train data  test data (7:3 )

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,  # 
    y,  # 
    test_size=0.3,  # 30%
    random_state=10)  #  

print('train data : ', len(X_train))

print('test data : ', len(X_test))
'''

Step 5:   - sklearn 

'''

# sklearn    

from sklearn.linear_model import LinearRegression

#    

lr = LinearRegression()

# train data   

lr.fit(X_train, y_train)

#    test data  (R-) 

r_square = lr.score(X_test, y_test)

print(r_square)

print('\n')

#  

print(' a: ', lr.coef_)

print('\n')

#  y

print('y b', lr.intercept_)

print('\n')

#   X     y_hat   y 

y_hat = lr.predict(X)

plt.figure(figsize=(10, 5))

ax1 = sns.distplot(y, hist=False, label="y")

ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1)

plt.show()

plt.close()
```
